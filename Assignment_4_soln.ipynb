```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plot a simple line graph using the values: x = [1, 2, 3, 4], y = [10, 20, 25, 30]

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)
plt.show()
```

```python
# Add a title and axis labels to the line graph in Question 1.

plt.title('line graph')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, y)
plt.show()
```

```python
# Plot a red dotted line for: x = [0, 1, 2, 3], y = [3, 6, 1, 8]

x = [0, 1, 2, 3]
y = [3, 6, 1, 8]

plt.plot(x, y, color = 'red', ls = ':')
plt.plot()
```

```python
# Plot two lines in the same graph: x = [1, 2, 3, 4], y1 = [10, 20, 25, 30], y2 = [5, 15, 20, 25]

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# Add a legend to the above plot to differentiate the two lines.

plt.plot(x, y1, label = 'first')
plt.plot(x, y2, label = 'second')
plt.legend()
plt.show()
```

```python
# Plot a bar chart for: Fruits = ['Apple', 'Banana', 'Mango'], Quantity = [10, 20, 15]

Fruits = ['Apple', 'Banana', 'Mango']
Quantity = [10, 20, 15]

plt.bar(Fruits, Quantity)
plt.show()
```

```python
# Plot a horizontal bar chart for the same fruit and quantity data
plt.barh(Fruits, Quantity)
```

```python
# Create a pie chart with percentage labels: Subjects = ['Math', 'Science', 'English'], Values = [30, 40, 30]

Subjects = ['Math', 'Science', 'English']
Values = [30, 40, 30]

plt.pie(Values, labels = Subjects)
plt.show()
```

```python
# Create a scatter plot: x = [5, 7, 8, 7], y = [8, 5, 6, 3]

x = [5, 7, 8, 7]
y = [8, 5, 6, 3]

plt.scatter(x, y)
plt.show()
```

```python
# Create a histogram for the data: marks = [60, 70, 80, 75, 70, 90, 85, 80]

marks = [60, 70, 80, 75, 70, 90, 85, 80]

plt.hist(marks, bins = 4, edgecolor = 'black')
plt.show()
```

```python
# Create a subplot with two plots: one line plot and one bar chart.

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))
ax1.plot(x, y)

ax2.bar(x, y)

plt.show()
```

```python
# Change the figure size to 8x5 while plotting a simple graph.

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.figure(figsize = (8, 5))
plt.show()
```

```python
# Plot a graph with customized line style, marker, and color.
# Save a plot as a PNG file using plt.savefig()
x = [1, 2, 3, 4]
y = [10, 22, 25, 12]

plt.plot(x, y, ls = "--", marker = 'o', color = 'black')
plt.savefig('hello.png')
plt.show()
```

```python
# Add grid lines to any plot you create.

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# Plot a line graph and annotate the highest point on it.

max_y = max(y)
max_x = x[y.index(max_y)]


plt.plot(x, y, marker = 'o', ls = '-')

plt.annotate(
    f'Highest Point\n({max_x}, {max_y})',
    xy = (max_x, max_y),
    xytext = (max_x + 0.5, max_y - 0.6),
    arrowprops = dict(facecolor = 'red', arrowstyle = '->'),
    color='red'
)


plt.grid(True)
plt.show()
```

```python
# Create a bar chart and add value labels on top of each bar.

fruits = ['Apples', 'Bananas', 'Cherries', 'Oranges', 'Grapes']
values = [25, 40, 30, 35, 20]

bars = plt.bar(fruits, values, color='skyblue')

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 1, f'{y}', ha = 'center', va = 'bottom')

plt.show()
```

```python
# Create a plot with x-axis rotated 45 degrees using tick labels: Months = ['Jan', 'Feb', 'Mar', 'Apr'], Sales = [250, 300, 150, 400]

months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [250, 300, 150, 400]

bars = plt.bar(months, sales, color='lightgreen')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f'{yval}', ha = 'center', va = 'bottom')

plt.xticks(rotation = 45)

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales ($)')

plt.show()
```

```python
# Draw three subplots in a single row, each with a different plot (line, bar, pie)

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (15, 5))
ax1.plot(x, y)

ax2.bar(x, y)

ax3.pie(x)

plt.show()
```

```python
# Use different colors for each bar in a bar chart using the color parameter

colors = ['red', 'blue', 'green', 'pink']
plt.bar(x, y, color = colors)
plt.show()
```
