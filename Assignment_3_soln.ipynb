```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, np.nan, 90, np.nan, 78]
})

print("Before filling missing values in 'Score':")
print(df)

df.loc[df['Score'].isnull(), 'Score'] = 75

print("\nAfter filling missing values with 75:")
print(df)
```

```python
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Math':    [85, 78, 92, 88, 76, 90],
    'Science': [89, 74, 95, 91, 80, 88],
    'English': [90, 82, 85, 87, 84, 89]
})

print("\nRows 2 to 5, Columns 1 to 3:")
print(df.iloc[2:6, 1:4])
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [19, 21, 20, 25, 18],
    'Grade': ['A', 'B', 'A', 'C', 'B']
})

filtered_df = df.loc[df['Age'] > 20]

print("\nRows where Age > 20:")
print(filtered_df)
```

```python
data = {
    'Item': ['Pen', 'Notebook', 'Eraser'],
    'Quantity': [10, 5, 15],
    'Price': [1.5, 3.0, 0.5]
}

items_df = pd.DataFrame(data)
items_df['Total'] = items_df['Quantity'] * items_df['Price']
print(items_df)
```

```python
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Temperature': [23.5, np.nan, 19.0, np.nan, 30.0]
}

df = pd.DataFrame(data)

print("Before filling missing values:")
print(df)

mean_temp = df['Temperature'].mean()

df.fillna({'Temperature': mean_temp}, inplace = True)

print("\nAfter filling missing values with the mean:")
print(df)
```
