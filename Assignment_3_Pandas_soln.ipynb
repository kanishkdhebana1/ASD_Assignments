```python
# Create a DataFrame with some NaN values in the Marks column. Replace all NaN values with the average (mean) of the Marks column.

df = pd.DataFrame({'Marks': [85, np.nan, 75, np.nan, 90]})
mean_val = df['Marks'].mean()
df
```

```python

df['Marks'] = df['Marks'].fillna(mean_val)
df = df.infer_objects(copy = False)
df
```

```python
# You are given a DataFrame df with missing values in the Score column. Use .loc to fill only those missing values with 75

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, np.nan, 90, np.nan, 78]
})

print("Before filling missing values in 'Score':")
print(df)

df.loc[df['Score'].isnull(), 'Score'] = 75

print("\nAfter filling missing values with 75:")
print(df)
```

```python
# Given a DataFrame df with student grades, use .iloc to print rows 2 to 5 and columns 1 to 3
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Math':    [85, 78, 92, 88, 76, 90],
    'Science': [89, 74, 95, 91, 80, 88],
    'English': [90, 82, 85, 87, 84, 89]
})

print("\nRows 2 to 5, Columns 1 to 3:")
print(df.iloc[2:6, 1:4])
```

```python
# Use .loc to filter and print all rows where the Age is greater than 20 from a DataFrame.

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [19, 21, 20, 25, 18],
    'Grade': ['A', 'B', 'A', 'C', 'B']
})

filtered_df = df.loc[df['Age'] > 20]

print("\nRows where Age > 20:")
print(filtered_df)
```

```python
# Create a DataFrame from a dictionary of items (Item, Quantity, Price) and calculate the total price for each row using a new column Total = Quantity * Price.

data = {
    'Item': ['Pen', 'Notebook', 'Eraser'],
    'Quantity': [10, 5, 15],
    'Price': [1.5, 3.0, 0.5]
}

items_df = pd.DataFrame(data)
items_df['Total'] = items_df['Quantity'] * items_df['Price']
print(items_df)
```

```python
#  You have a DataFrame of city temperatures. Some values in the Temperature column are missing. Fill them using the average temperature calculated using mean().
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Temperature': [23.5, np.nan, 19.0, np.nan, 30.0]
}

df = pd.DataFrame(data)

print("Before filling missing values:")
print(df)

mean_temp = df['Temperature'].mean()

df.fillna({'Temperature': mean_temp}, inplace = True)

print("\nAfter filling missing values with the mean:")
print(df)
```
