```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the 'tips' dataset from Seaborn and display the first 5 row.

tips = sns.load_dataset('tips')
tips.head(5)
```

```python
# Create a scatter plot of total_bill vs tip using the 'tips' dataset.

sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'tip')
plt.show()
```

```python
# Plot a line graph for total_bill vs tip using Seaborn.

sns.lineplot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
# Draw a bar plot to show the average tip for each day in the 'tips' dataset.

sns.barplot(data = tips, x = 'day', y = 'tip', hue = 'day')
plt.show()
```

```python
# Use a count plot to display how many records are available for each day.

sns.countplot(data = tips, x = 'day', hue = 'day')
plt.show()
```

```python
# Make a box plot to show total_bill distribution across different days.

sns.boxplot(data = tips, x = 'total_bill', y = 'day', hue = 'day')
plt.show()
```

```python
# Create a violin plot for total_bill vs day.

sns.violinplot(data = tips, x = 'total_bill', y = 'day', hue = 'day')
```

```python
# Plot a histogram of the 'total_bill' column with KDE using histplot.

sns.histplot(data = tips, x = 'total_bill', kde = True, color = 'green')
plt.plot()
```

```python
# Use the 'titanic' dataset and show a heatmap of its numeric correlation matrix.

titanic = sns.load_dataset('titanic')
titanic.head(5)
```

```python
corr = titanic.corr(numeric_only = True)
sns.heatmap(data = corr, annot = True)
plt.show()
```

```python
#Convert all categorical columns in the 'tips' dataset to numeric and create a full correlation heatmap.

tips_copy = tips.copy()
tips_copy.head()
```

```python
columns = ['sex', 'smoker', 'day', 'time']
for x in columns:
    print(f"Unique values for {x}: {tips[x].unique()}")
```

```python
tips_copy['sex'] = tips_copy['sex'].map({'Male': 0, 'Female': 1})
tips_copy['smoker'] = tips_copy['smoker'].map({'No': 0, 'Yes': 1})
tips_copy['day'] = tips_copy['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
tips_copy['time'] = tips_copy['time'].map({'Lunch': 0, 'Dinner': 1})
```

```python
tips_copy
```

```python
corr = tips_copy.corr()
sns.heatmap(data = corr, annot = True)
plt.show()
```
